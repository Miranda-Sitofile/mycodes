```python
##Exercise 4: Plot a time series of the global sum of temperature and dissolved meadian for the period 1850 - 2005
```

```python
def F_time_series_anomalies(var,area):
    """computes time series anamolies"""
    
    ## calculate the anomaly by removing the mean 
    var_anomaly = var - var.mean('year')
    
    ## compute total area sum
    area_sum_ds = area.sum(dim = ['nlon','nlat'])
    
    ## calculate the weighted mean
    var_anomaly_ds = ((var_anomaly*area).sum(dim = ['nlon','nlat']))/area_sum_ds
    
    return var_anomaly_ds.values
    
    
temp_200m_amom_ds = F_time_series_anomalies(cesm_temp_200m_ds.TEMP,area_ds)
o2_200m_amom_ds = F_time_series_anomalies(cesm_o2_200m_ds.O2,area_ds)
```

```python
##Exercise 5: calculate the correlation coeffient between Temperature and dissolved oxygen global means for period 1850 - 1960 and 1960 - 2005
```

```python
time_1850_1960 = slice('1850', '1960')
time_1960_2005 = slice('1960', '2005')
cesm_temp_200m_1850_1960_ds = cesm_temp_200m_ds.TEMP.sel(year = time_1850_1960)
cesm_temp_200m_1960_2005_ds = cesm_temp_200m_ds.TEMP.sel(year = time_1960_2005)

cesm_o2_200m_1850_1960_ds = cesm_o2_200m_ds.O2.sel(year = time_1850_1960)
cesm_o2_200m_1960_2005_ds = cesm_o2_200m_ds.O2.sel(year = time_1960_2005)

cesm_temp_1850_1960_ts = (((cesm_temp_200m_1850_1960_ds*area_ds.TAREA).sum(dim =\
                                                                     ['nlat','nlon']))/(area_ds.TAREA.sum(dim = ['nlat','nlon'])))
cesm_o2_1850_1960_ts = (((cesm_o2_200m_1850_1960_ds*area_ds.TAREA).sum(dim =\
                                                                     ['nlat','nlon']))/(area_ds.TAREA.sum(dim = ['nlat','nlon'])))
 
cesm_temp_1960_2005_ts = (((cesm_temp_200m_1960_2005_ds*area_ds.TAREA).sum(dim =\
  
                                                                    ['nlat','nlon']))/(area_ds.TAREA.sum(dim = ['nlat','nlon'])))

cesm_o2_1960_2005_ts = (((cesm_o2_200m_1960_2005_ds*area_ds.TAREA).sum(dim =\
                                                                     ['nlat','nlon']))/(area_ds.TAREA.sum(dim = ['nlat','nlon'])))  
```

```python
print(np.corrcoef(cesm_temp_1850_1960_ts,cesm_o2_1850_1960_ts)[1,0])
print(np.corrcoef(cesm_temp_1960_2005_ts,cesm_o2_1960_2005_ts)[1,0])
```
